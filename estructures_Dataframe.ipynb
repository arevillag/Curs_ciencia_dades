```python
# Exercici 1
# Descarrega el data set Airlines Delay: Airline on-time statistics and delay causes i carrega’l a un pandas Dataframe.
# Llibreria pandas i carreguem les dades dels Arlines Delay que estan en format CSV
import pandas as pd

df = pd.read_csv('archive/DelayedFlights.csv', header=0, sep=',',encoding='latin1', dayfirst=True)

# Visualitzem les dades sense filtre
df.shape
df.info()
print("Numeró de dades es de 1.936.758 dades dels vols i tenim 29 variables.") 

# Filtrem 100 registres y creem dfFiltrat
dfFiltrat = df[:100] 
print (dfFiltrat)

print ("Dades sumades df dfFiltrat \n")
print(dfFiltrat.isnull().sum())
print ("Dades sumades df original \n")
print(df.isnull().sum())

# Les variables mes rellevant son les que están relacionades amb el temps de retràs i arribada , la variable Year sempre te el mateix valor la treiem de les dades.
# Dades rellevants que proporciona describe.
df.describe()

# Treiem variable Year del conjunt de dades
df2 = df.loc[:, df.columns != 'Year']

# Ara quedaria la capçalera de aquesta forma.
print ("Dades capçalera sensa columna Year \n")
df2.describe()
```

```python
"""Exercici 2
Resumeix estadísticament les columnes d’interès
Troba quantes dades faltants hi ha per columna
Crea columnes noves (velocitat mitjana del vol, si ha arribat tard o no...)
Taula de les aerolínies amb més endarreriments acumulats
Quins són els vols més llargs? I els més endarrerits?"""

#Resumeix estadísticament les columnes d’interès
print("Resumeix estadísticament les columnes d’interès \n" +  str(df2.describe()))

# Analitzem les variables que són categòriques per poder clasificarles
print("Variables que són categòriques per mes \n")
mes_counts = df2['Month'].value_counts(sort = False)
mes_counts.index = ('Gener','Febrer','Març','Abril','Maig','Juny','Juliol','Agost','Setembre','Octubre','Novembre','Desembre')
# Visaulitzem grafic amb barras
mes_counts.plot(kind='bar')

print("Variables que són categòriques per dia \n")
dia_counts = df2['DayOfWeek'].value_counts(sort = False)
dia_counts.index = ('Dll','Dm','Dx','Dj','Dv','Ds','Dg')
# Visaulitzem grafic amb barras
dia_counts.plot(kind = 'bar')

# LAX és l'aeroport amb menys sortides
# Extraiem dades i conclusions de la variable Origin
print ("Dades de la variable Origen: \n " + str(df2['Origin'].value_counts()))

# ATL és l'aeroport amb mes arribades
# Extraiem dades i conclusions de la variable Origin
print ("Dades de la variable Desti: \n" + str(df2['Dest'].value_counts()))

# lliste de companyies amb més vols la primera es WN i la últioma AQ
print ("Dades de la variable UniqueCarrier\n" + str(df2['UniqueCarrier'].value_counts()))

# Dades que son missing per variable
# Les variables que son missings estan relacionades amb els vols amb retràs.
print("Dades que son missing per variable \n" + str(df2.isnull().sum()))

#Creem variables noves (velocitat mitjana del vol, si ha arribat tard o no...)
# Extraiem dades i conclusions de la variable nova
# La Velocita mmes alta en el viatges es 6253
df2['Velocitat'] = round(df2['Distance'] / df2['ActualElapsedTime'],2)
print ("Dades de la nova variable Velocitat\n" + str(df2['Velocitat'].value_counts()))


# Extraiem dades i conclusions de la variable nova
# La ruta que te menys viatges es LAS-LAX
df2['Ruta'] = df2['Origin'] + "-" + df2['Dest']
print ("Dades de la nova variable Ruta\n" + str(df2['Ruta'].value_counts()))

# la ruta que té més distància es de 4962.
print("Quins són els vols més llargs \n" + str(df2["Distance"].max()))
df2[df2['Distance'] == df2['Distance'].max()].groupby('Ruta').first()
print(df2.groupby('Ruta')['Velocitat'].mean())


# Analitzem els vols amb retràs
# Extraiem dades i conclusions de la variable nova
# Molt vols han tingut retras
print("Vols amb més retras \n")
df2['retras'] = pd.notnull(df["CarrierDelay"])
print (df2['retras'].value_counts())

#Creem sub de dades
df2.delay = df2[pd.notnull(df2["CarrierDelay"])]
df2.delay.shape

# La ruta amb més retras es LAX-SFO i la que menys RDD-SMF
print("Rutes amb més retras \n")
df2.delay['Ruta'].value_counts()
```
