```python
#Exercici 2
import pandas as pd
import numpy as np

df = pd.read_csv('archive/DelayedFlights.csv', header=0, sep=',',encoding='latin1', dayfirst=True)

# Columnes rellevants.
dfFiltrat = df[['FlightNum','Year','Month','DayofMonth','DepTime','ArrTime','TailNum','ActualElapsedTime','AirTime','ArrDelay','DepDelay','Origin','Dest','Distance','Cancelled','CancellationCode']]

#Resumeix estadísticament les columnes d’interès
print("Resumeix estadísticament les columnes d’interès \n" +  str(dfFiltrat.describe()))
#Troba quantes dades faltants hi ha per columna
print("Troba quantes dades faltants hi ha per columna \n" + str(dfFiltrat.isnull().sum()))
#Crea columnes noves (velocitat mitjana del vol, si ha arribat tard o no...)
dfFiltrat = dfFiltrat.assign(mitjanaVol=np.mean(dfFiltrat['ActualElapsedTime'],axis=0), arribatTard='N')
print("Crea columnes noves (velocitat mitjana del vol, si ha arribat tard o no...) \n" + str(dfFiltrat))
#Taula de les aerolínies amb més endarreriments acumulats
#print(dfFiltrat.groupby(level="DepDelay").mean())
#Quins són els vols més llargs
print("Quins són els vols més llargs \n" + str(dfFiltrat["Distance"].max()))
```
