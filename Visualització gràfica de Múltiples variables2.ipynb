```python
# Create data
x = movies.movie_title
y = movies.genres
colors = 'Blue'
area = np.pi*5

plt.axis([-10, 100, 0, 60])

# Plot
plt.scatter(x, y, s=area, c=colors, alpha=0.4, edgecolors='black')

plt.title('Movie title vs genres')
plt.ylabel('Movie title)
plt.xlabel('genres')
plt.show()
```

```python
fig = plt.figure(figsize=(15,4))

ax1 = fig.add_subplot(1, 3, 1)

# Create data
x = movies.movie_title
y = movies.genres
colors = 'Blue'
area = np.pi*5

plt.axis([0, 100, 0, 60])

# Plot
ax1.scatter(x, y, s=area, c=colors, alpha=0.4, edgecolors='black')

plt.title('Movies Dataset')
plt.ylabel('Movies title')
plt.xlabel('Genres')

ax2 = fig.add_subplot(1, 3, 2)
# Create data
x = movies.movie_title
y = movies.genres

colors = "Red"
area = np.pi*20
plt.axis([0, 70, 0, 1200])

# Plot
ax2.scatter(x, y, s=area, c=colors, alpha=0.4, edgecolors='black')
plt.title('Movies Dataset 2')
plt.ylabel('Movies title')
plt.xlabel('Genres')

ax3 = fig.add_subplot(1, 3, 3)
# Create data
x = movies.movie_title
y = movies.genres

colors = "Orange"
area = np.pi*30
plt.axis([1500, 4300, 0, 65])

# Plot
ax3.scatter(x, y, s=area, c=colors, alpha=0.4, edgecolors='black')
plt.title('Movies Dataset 3')
plt.ylabel('Movies title')
plt.xlabel('Genres')


plt.tight_layout()
```
