```python
# Modul 2
# Exercici 1
# Resumeix gràficament el data set DelayedFlights.csv
# Una variable categòrica (UniqueCarrier)
# Una variable numèrica (ArrDelay)
# Una variable numèrica i una categòrica (ArrDelay i UniqueCarrier)
# Dues variables numèriques (ArrDelay i DepDelay)
# Tres variables (ArrDelay, DepDelay i UniqueCarrier)
# Més de tres variables (ArrDelay, DepDelay, AirTime i UniqueCarrier).

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('archive/DelayedFlights.csv', header=0, sep=',',encoding='latin1', dayfirst=True)

# Una variable categòrica (UniqueCarrier)
# Diagrama de barres, cambia el color de les barres i afegim noms als eixos i titol
carrier_counts = df['UniqueCarrier'].value_counts()
plt.bar(carrier_counts.index,carrier_counts.values,color='C7')
plt.xlabel("Transport Unic")
plt.ylabel("Numero de vols")
plt.title("Vols")
plt.show()

# Una variable numèrica (ArrDelay)
# Diagrama de barres, cambia el color de les barres i afegim noms als eixos i titol
plt.hist(df['ArrDelay'],color="yellow")
plt.xlabel("Arribades amb retart")
plt.title("Arribades")
plt.show()

# Revisem les dades
# BoxPlot afegim noms als eixos i titol
print(df['ArrDelay'].describe())
ax = sns.boxplot(x=df['ArrDelay'])
plt.xlabel("Arribades amb retart")
plt.show()

# Una variable numèrica i una categòrica (ArrDelay i UniqueCarrier)
# BoxPlot afegim noms als eixos i titol
ax = sns.boxplot(x=df['ArrDelay'],y=df['UniqueCarrier'])
plt.xlabel("Arribades amb retart")
plt.ylabel("Transport Unic")
plt.title("Grafic Arribades retart i transport unic")
plt.show()

# Visualització de dues variables numèriques (ArrDelay i DepDelay)
# Grafic afegim noms als eixos i titol
plt.scatter(df['ArrDelay'],df['DepDelay'],marker='.')
plt.xlabel("Arribades amb retart")
plt.ylabel("Sortides amb retart")
plt.title(label="Grafic dues variables", fontsize=20, color="grey")
plt.show()

# Visualització de més de tres variables (ArrDelay, DepDelay, AirTime i UniqueCarrier)
# Grafic afegim noms als eixos i titol
fig, axs = plt.subplots(1, 3, figsize=(15,7))
############## 3 variables ##############
sns.boxplot(x=df['ArrDelay'],y=df['UniqueCarrier'],ax=axs[0])
sns.boxplot(x=df['DepDelay'],y=df['UniqueCarrier'],ax=axs[1])
sns.boxplot(x=df['AirTime'],y=df['UniqueCarrier'],ax=axs[2])
############## 3 etiquetas ##############
axs[0].set_xlabel('Arribades amb retart')
axs[1].set_xlabel('Sortides amb retart')
axs[2].set_xlabel('Temps vol')
#########################################
plt.title(label="Grafic tres variables", fontsize=20, color="blue")
plt.show()
```

```python
# Modul 2
# Exercici 2
# Exporta els gràfics com imatges o com html.

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('archive/DelayedFlights.csv', header=0, sep=',',encoding='latin1', dayfirst=True)

# Grafics guardats en HTML

# Una variable categòrica (UniqueCarrier)
# Diagrama de barres, cambia el color de les barres i afegim noms als eixos i titol
carrier_counts = df['UniqueCarrier'].value_counts()
plt.bar(carrier_counts.index,carrier_counts.values,color='C7')
plt.xlabel("Transport Unic")
plt.ylabel("Numero de vols")
plt.title("Vols")
plt.savefig("UniqueCarrier.png") # Guardem el grafic en format png
plt.show()

# Creem un fitxer en format html i despres incrustem la imatge que em creat en format imatge
f = open('UniqueCarrier.html','w')
mensaje= "Content-type: text/html\n"
mensaje+= """<html>
                <body>
                    <img src="UniqueCarrier.png"></img>
                </body>
             </html>"""
f.write(mensaje)
f.close()

print("Fitxer UniqueCarrier.png creat en format html\n")


# Una variable numèrica (ArrDelay)
# Diagrama de barres, cambia el color de les barres i afegim noms als eixos i titol
plt.hist(df['ArrDelay'],color="yellow")
plt.xlabel("Arribades amb retart")
plt.title("Arribades")
plt.savefig("VariableNumerica.png") #Guardem grafic
plt.show()

# Creem un fitxer en format html i despres incrustem la imatge que em creat en format imatge
f = open('UniqueCarrier.html','w')
mensaje= "Content-type: text/html\n"
mensaje+= """<html>
                <body>
                    <img src="VariableNumerica.png"></img>
                </body>
             </html>"""
f.write(mensaje)
f.close()

print("Fitxer VariableNumerica.png creat en format html\n")


# Grafics en imatge png

# Revisem les dades
# BoxPlot afegim noms als eixos i titol
print(df['ArrDelay'].describe())
ax = sns.boxplot(x=df['ArrDelay'])
plt.xlabel("Arribades amb retart")
plt.savefig("Grafic Arribades amb retart.png")
plt.show()

# Una variable numèrica i una categòrica (ArrDelay i UniqueCarrier)
# BoxPlot afegim noms als eixos i titol
ax = sns.boxplot(x=df['ArrDelay'],y=df['UniqueCarrier'])
plt.xlabel("Arribades amb retart")
plt.ylabel("Transport Unic")
plt.title("Grafic Arribades retart i transport unic")
plt.savefig("Grafic ArrDelay-UniqueCarrier.png") #Guardem grafic
plt.show()

# Visualització de dues variables numèriques (ArrDelay i DepDelay)
# Grafic afegim noms als eixos i titol
plt.scatter(df['ArrDelay'],df['DepDelay'],marker='.')
plt.xlabel("Arribades amb retart")
plt.ylabel("Sortides amb retart")
plt.title(label="Grafic dues variables", fontsize=20, color="grey")
plt.savefig("Grafic retart.png") #Guardem grafic
plt.show()

# Visualització de més de tres variables (ArrDelay, DepDelay, AirTime i UniqueCarrier)
# Grafic afegim noms als eixos i titol
fig, axs = plt.subplots(1, 3, figsize=(15,7))
############## 3 variables ##############
sns.boxplot(x=df['ArrDelay'],y=df['UniqueCarrier'],ax=axs[0])
sns.boxplot(x=df['DepDelay'],y=df['UniqueCarrier'],ax=axs[1])
sns.boxplot(x=df['AirTime'],y=df['UniqueCarrier'],ax=axs[2])
############## 3 etiquetas ##############
axs[0].set_xlabel('Arribades amb retart')
axs[1].set_xlabel('Sortides amb retart')
axs[2].set_xlabel('Temps vol')
#########################################
plt.title(label="Grafic tres variables", fontsize=20, color="blue")
plt.savefig("Grafic Delay-AirTime-UniqueCarrier.png") #Guardem grafic
plt.show()
```

```python
# Modul 2
# Exercici 3
# Integra les visualitzacions gràfiques, en la tasca 2, del Mòdul 1.

# Número de vols per mes
mes_counts = df['Month'].value_counts(sort = False)
mes_counts.index = ('Gener','Febrer','Març','Abril','Maig','Juny','Juliol','Agost','Setembre','Octubre','Novembre','Desembre')

# Visualització els vols mensuals
plt.bar(mes_counts.index,mes_counts.values,width=0.7,color='Red')
plt.xlabel("Mesos")
plt.ylabel("Numero de vols")
plt.title("Vols mensuals")
plt.show()

# Número de vols diaris
dia_counts = df['DayOfWeek'].value_counts(sort = False)
dia_counts.index = ('Dll','Dm','Dx','Dj','Dv','Ds','Dg')

# Visualització els vols diaris
plt.bar(dia_counts.index,dia_counts.values,width=0.7,color='Black')
plt.xlabel("Dia")
plt.ylabel("Numero de vols")
plt.title("Vols diaris")
plt.show()
```
