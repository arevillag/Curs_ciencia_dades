```python
# Modul 2
# Exercici 1
# Resumeix gr√†ficament el data set DelayedFlights.csv
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('archive/DelayedFlights.csv', header=0, sep=',',encoding='latin1', dayfirst=True)
# Columnes rellevants.
dfFiltrat = df[['UniqueCarrier','ArrDelay','FlightNum','Year','Month','DayofMonth','DepTime','ArrTime','TailNum','ActualElapsedTime','AirTime','DepDelay','Origin','Dest','Distance','Cancelled','CancellationCode']]


# Una variable categ√≤rica (UniqueCarrier)
x_values = dfFiltrat['UniqueCarrier'].unique()
y_values = dfFiltrat['UniqueCarrier'].value_counts().tolist()
plt.bar(x_values, y_values)
plt.title("Una variable categ√≤rica (UniqueCarrier)")
plt.xlabel("(Tipus)")
plt.ylabel("(Quantitat)")
plt.show()
plt.close('all')


# Una variable num√®rica (ArrDelay)
x_values = dfFiltrat['ArrDelay'].value_counts().tolist()
y_values = dfFiltrat['ArrDelay'].value_counts().tolist()
plt.scatter(x_values, y_values)
plt.title("Una variable num√®rica (ArrDelay)")
plt.xlabel("(Valor)")
plt.ylabel("(Quantitat)")
plt.show()
plt.close('all')

# Dues variables num√®riques (ArrDelay i DepDelay)
x = dfFiltrat['DepDelay'].value_counts().tolist()
y = dfFiltrat['DepDelay'].value_counts().tolist()
x2 = dfFiltrat['ArrDelay'].value_counts().tolist()
y2 = dfFiltrat['ArrDelay'].value_counts().tolist()

plt.plot(x, y, label='DepDelay')
plt.plot(x2, y2, label='ArrDelay')

plt.title("Dues variables num√®riques (ArrDelay i DepDelay)")
plt.grid(True)
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
plt.close('all')
```

```python
# Modul 2
# Exercici 2
# Exporta els gr√†fics com imatges o com html.

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('archive/DelayedFlights.csv', header=0, sep=',',encoding='latin1', dayfirst=True)
# Columnes rellevants.
dfFiltrat = df[['UniqueCarrier','ArrDelay','FlightNum','Year','Month','DayofMonth','DepTime','ArrTime','TailNum','ActualElapsedTime','AirTime','DepDelay','Origin','Dest','Distance','Cancelled','CancellationCode']]


# Una variable categ√≤rica (UniqueCarrier)
x_values = dfFiltrat['UniqueCarrier'].unique()
y_values = dfFiltrat['UniqueCarrier'].value_counts().tolist()
plt.bar(x_values, y_values)
plt.title("Una variable categ√≤rica (UniqueCarrier)")
plt.xlabel("(Tipus)")
plt.ylabel("(Quantitat)")

# Guardem el grafic en format png
plt.savefig('UniqueCarrier.png')
print("Fitxer UniqueCarrier.png creat en format imatge\n")


# Creem un fitxer en format html i despres incrustem la imatge que em creat en format imatge
f = open('UniqueCarrier.html','w')
mensaje= "Content-type: text/html\n"
mensaje+= """<html>
                <body>
                    <img src="UniqueCarrier.png"></img>
                </body>
             </html>"""
f.write(mensaje)
f.close()

print("Fitxer UniqueCarrier.png creat en format html\n")
```
